```r
library(tidyverse)
```

```r
data = read.csv("/Users/jessicakaut/Downloads/Car-Sales-Data.csv")
```

```r
head(data)
```

```r
model <- lm(Price ~ Brand + Body + Mileage + EngineV + Engine.Type + Registration + Year + Model, data = data)
```

```r
summary(model)
```

# Examine F-statistic

## Since the p-value is way below .001 (2.2e-16) the effect of the specifications are extremely significant on the price.

# Which predictor variables are significant?

```r
summary(model)$coefficient
```

## Looks like the registration status (registrationyes) and Year are significantly associated with price.

# Confidence Interval

```r
confint(model)
```

# Examining R2 value

## Adjusted R2 = 0.74, meaning that 74% of the variance in the measure of price can be predicted by the car specifications.

# Hybrid stepwise

```r
fitstart = lm(Price ~ 1, data = data)
```

```r
Fitall = lm(Price ~ ., data = data)
```

```r
step(fitstart, direction="both", scope=formula(FitAll))
```

## Using the AIC values above: The best way to improve the above model is to add model and year.
